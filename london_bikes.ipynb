```python
# Specifying Column Names that I want to use.
new_cols_dict = {
    'timestamp' : 'time',
    'cnt' : 'count',
    't1' : 'temperature',
    't2' : 'temperature_feels_like_C',
    'hum' : 'humidity_percent',
    'wind_speed' : 'wind_speed_kph',
    'weather_code' : 'weather',
}

# Renaming the columns to the specified column names
london_bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
#  Changing the humidity values to percentage
london_bikes.humidity_percent = london_bikes.humidity_percent / 100
```

```python
#  Creating Season Dict
season_dict = {
    '0.0' : 'Spring',
    '1.0' : 'Summer',
    '2.0' : 'Fall',
    '3.0' : 'Winter'
}

# Changing Season Data type from Float to String, and mapping the values.
london_bikes.season = london_bikes.season.astype('str')
london_bikes.season = london_bikes.season.map(season_dict)

# Creating Weather Dict
weather_dict = {
    '1.0' : 'Clear',
    '2.0' : 'Scattered Clouds',
    '3.0' : 'Broken Clouds',
    '4.0' : 'Cloudy',
    '7.0' : 'Rain / Light Rain Shower / Light Rain',
    '10.0' : 'Rain With Thunderstorm',
    '26.0' : 'Snowfall',
    '94.0' : 'Freezing Fog'
}

# Changing Weather Data Type From Float to String and mapping the values.
london_bikes.weather = london_bikes.weather.astype('str')
london_bikes.weather = london_bikes.weather.map(weather_dict)
```
